# How to find out all missing values for each columns?

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.describe().round(2)
```

```python
# By understanding the data in a more descriptive manner,
# we notice one thing:
# Quantity has negative values, 
# At this stage, we’ll just remove Quantity with negative values.
```

Quantity has negative values records

```python
df[df.Quantity<0]
```

```python
df[df.Quantity<0].shape
```

How to remove all Quantity records with negative values?

```python
df_new=df[~(df['Quantity']<0)]
```

```python
df_new
```

```python
df_new.shape
```

```python
# To calculate the total money spent on each purchase, 
# we simply multiply Quantity with Unit Price:
# amount_spent = quantity * unit_price
```

```python
amount_spent=df_new.Quantity*df_new.UnitPrice
```

```python
df_new.insert(6,'Amount_Spent',amount_spent)
```

```python
df_new
```

Finally, we add a few columns that consist of the Year,
Month,Day and Hour for Each transaction for analysis later

```python
df_new.dtypes
```

```python
df_new['InvoiceDate']=pd.to_datetime(df_new.InvoiceDate)
```

```python
df_new.dtypes
```

```python
df_new
```

```python
df_new.insert(5,'Year',df_new.InvoiceDate.dt.year)
```

```python
df_new
```

```python
df_new.insert(6,'Month',df_new.InvoiceDate.dt.month)
```

```python
df_new
```

```python
df_new.insert(7,'Day',df_new.InvoiceDate.dt.day)
```

```python
df_new
```

```python
df_new.insert(8,'Hour',df_new.InvoiceDate.dt.hour)
```

```python
df_new
```

The Final Dataframe Look Like This:-

```python
df_new
```

```python
df_new=df_new.assign(Year=df_new.InvoiceDate.dt.year,
                     Day=df_new.InvoiceDate.dt.day
                    Hour=df_new.InvoiceDate.dt.hour)
```

Top 5 Customers with Highest Money Spent

```python
df_n=df_new.Amount_Spent.sort_values(ascending=False).index[:5]
df_n
```

```python
df_new.loc[df_n,['CustomerID','Country','Amount_Spent']]
```

```python
import matplotlib.pyplot as plt
```

How Many Orders (per Month)?

```python
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.title('Number of Orders of Different Months')
df_new.Month.value_counts().plot(kind='bar')
plt.show()
```

```python
plt.xlabel('Day')
plt.ylabel('Number of Orders')
plt.title('Number of Orders of Different Days')
df_new.Day.value_counts().plot(kind='bar')
plt.show()
```

```python
plt.xlabel('Hours')
plt.ylabel('Number of Orders')
plt.title('Number of Orders of Different Hours')
df_new.Hour.value_counts().plot(kind='bar')
plt.show()
```
